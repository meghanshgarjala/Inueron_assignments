Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a Python visualization library based on Matplotlib, offering a higher-level interface for drawing attractive statistical graphics. Here are five plots commonly used with Seaborn:
    
Scatter Plot
Use: Visualize the relationship between two variables, showing individual data points.
    
Line Plot
Use: Display trends over time or any other continuous variable.
    
Bar Plot
Use: Compare categorical data across different groups.

Box Plot
Use: Summarize the distribution of data and detect outliers.
    
Histogram
Use: Show the distribution of a single numerical variable.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
plt.title('Signal vs Timepoint for different Events and Regions (fmri dataset)')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
# Load titanic dataset
titanic = sns.load_dataset("titanic")

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Age distribution across Passenger Class')
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Fare distribution across Passenger Class')
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', kde=True)
plt.title('Histogram of Price with Cut Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
iris = sns.load_dataset("iris")

sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset("flights")

flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap='coolwarm', linecolor='white', linewidth=1)
plt.title('Passengers Heatmap (Flights Dataset)')
plt.show()
```
